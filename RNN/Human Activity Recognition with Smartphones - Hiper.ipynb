```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.model_selection import GridSearchCV

from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
df_train = pd.read_csv("C:\\AmaruAyelen\\proyectos\\PythonDataScienceMachineLearning\\PythonDataScienceMachineLearning\\bootcampIA\\RNN\\train.csv")
df_train.head()
```

```python
df_train.dtypes
```

```python
df_test = pd.read_csv("C:\\AmaruAyelen\\proyectos\\PythonDataScienceMachineLearning\\PythonDataScienceMachineLearning\\bootcampIA\\RNN\\test.csv")
df_test.head()
```

```python
# Separar las características (X) de la etiqueta (y)
X_train = df_train.drop('Activity', axis=1)
y_train = df_train['Activity']

X_test = df_test.drop('Activity', axis=1)
y_test = df_test['Activity']
```

```python
# Revisar las dimensiones
print(f"Características (X): {X_train.shape}")
print(f"Etiquetas (y): {y_train.shape}")
```

```python
# Normalizar las características usando StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
# Ver la forma de los datos
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de y_train: {y_train.shape}")
print(f"Forma de X_test: {X_test.shape}")
print(f"Forma de y_test: {y_test.shape}")
```

```python
y_test.head()
```

```python
# Convertir y_train y y_test de texto a valores numéricos
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Mostrar las etiquetas originales y las codificadas
print(f"y_train: {list(y_train[:5])}")
print(f"y_test: {y_test[:5]}")
```

```python
# definir time_steps
num_filas = X_train.shape[1]
time_steps = num_filas
time_steps
```

```python
# Definir el número de clases
num_classes = len(np.unique(y_train))
```

```python
# Remodelar los datos: SimpleRNN espera que los datos estén en forma (samples, timesteps, features)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
model = Sequential()
model.add(SimpleRNN(units=100, activation='tanh', input_shape=(time_steps, 1), return_sequences=True)) 
model.add(SimpleRNN(units=50, activation='relu'))  # Segunda capa LSTM sin return_sequences
model.add(Dense(num_classes, activation='softmax'))  # Capa de salida
```

```python
# Compilación del modelo
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Implementación Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, min_lr=1e-6)
```

```python
# Entrenamiento del modelo
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test), 
                    callbacks=[early_stopping, reduce_lr])
```

```python
# Definir la grilla de hiperparámetros
param_grid = {
    'units': [50, 100],  # Número de unidades en la capa RNN
    'activation': ['tanh', 'relu'],  # Función de activación
    'optimizer': ['adam', 'sgd'],  # Optimizadores
    'epochs': [10, 20],  # Número de épocas
    'batch_size': [64, 128]  # Tamaños de batch
}
```

```python
# Crear el objeto GridSearchCV con validación cruzada
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=1)
```

```python
# Ajustar la búsqueda de hiperparámetros con los datos de entrenamiento
grid_result = grid.fit(X_train, y_train, callbacks=[early_stopping], validation_data=(X_test, y_test))
```

```python
# Imprimir los mejores hiperparámetros
print(f"Mejores hiperparámetros: {grid_result.best_params_}")
```

```python
# Evaluar el mejor modelo en el conjunto de prueba
accuracy = grid_result.score(X_test, y_test)
print(f"Accuracy del mejor modelo en el conjunto de prueba: {accuracy}")
```

# ESTOS SON LOS RESULTADOS DE LAS MÉTRICAS ANTES DE APLICAR BÚSQUEDA DE HIPERPARÁMETROS Y BÚSQUEDA POR GRILLA

```python
# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test),
                    callbacks=[early_stopping])
```

```python
# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Loss en el conjunto de prueba: {loss}")
print(f"Accuracy en el conjunto de prueba: {accuracy}")
```

```python
# Hacer predicciones
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
```

```python
# Mostrar un reporte de clasificación
print(classification_report(y_test, y_pred_classes))
```
