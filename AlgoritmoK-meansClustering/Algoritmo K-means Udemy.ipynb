```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
# Establecer la semilla aleatoria para la reproducibilidad
np.random.seed(42)

# Generar un conjunto de datos de 100 puntos con 2 dimensiones cada uno
datos = np.random.rand(100, 2)
```

```python
# Indiccaión cantidad de clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(datos)
```

```python
# Almacena las coordenadas de los centroides después de que el algoritmo ha sido entrenado en el conjunto de datos.
centroides = kmeans.cluster_centers_
centroides
```

```python
# Se asigna etiqueta para cada fila de 0 a 4
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(datos[:, 0], datos[:, 1], c= etiquetas, cmap='viridis', marker='o');
plt.scatter(centroides[:,0],centroides[:,1], c='red', marker='X', s=200, label='Centroides');
plt.title('Visualización de K-Means Clustering')
```

```python
# Visualizar los datos
plt.scatter(datos[:, 0], datos[:, 1])
plt.title('Datos aleatorios con valores entre 0 y 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Ejercicio con los pinguinos

```python
pinguinos = sns.load_dataset('penguins')
```

```python
pinguinos.dropna(inplace=True)
```

```python
pinguinos.head()
```

```python
data = pinguinos[['bill_length_mm','bill_depth_mm']]
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(data)
```

```python
centroides = kmeans.cluster_centers_
etiquetas1 = kmeans.labels_
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=pinguinos,x='bill_length_mm',y='bill_depth_mm',hue=etiquetas1,palette='viridis')
plt.scatter(centroides[:,0], centroides[:,1], c='red', s=100, marker='*', label='Centroides')
plt.title('K-meas CLustering en el conjunto de datos Pinguis')
plt.legend()
plt.xlabel=('Longitud Pico (mm)')
plt.ylabel=('Profundidad Pico (mm)');
```
