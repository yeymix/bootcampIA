```python
import numpy as np # type: ignore 
import pandas as pd # type: ignore 
import matplotlib.pyplot as plt # type: ignore 
from sklearn.cluster import KMeans # type: ignore
import seaborn as sns # type: ignore 
```

```python
# Establecer la semilla aleatoria para la reproducibilidad
np.random.seed(42)

# Generar un conjunto de datos de 100 puntos con 2 dimensiones cada uno
datos = np.random.rand(100, 2)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(datos)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(datos)
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
etiquetas = kmeans.labels_
etiquetas
```

```python
plt.figure(figsize=(10,6))
plt.scatter(datos[:,0],datos[:,1],c=etiquetas,cmap='viridis',marker='o')
plt.scatter(centroides[:,0],centroides[:,1],c='red',marker='x',s=200, label='Centroides')
plt.legend()
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
```
