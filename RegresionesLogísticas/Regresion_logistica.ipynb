```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# División conjunto de entrenamiento
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # Biblioteca para usar regresiones logísticas
from sklearn.metrics import accuracy_score,confusion_matrix # librería necesaria para calcular la matriz de confusión.
```

```python
data = {
    'Horas_estudiadas':[1,2,3,4,5,6,7,8,9],
    'Horas_sueno': [7,6,8,5,9,7,8,6,5],
    'Aprobo':[0,0,0,1,1,1,1,1,1]
}

df = pd.DataFrame(data)

X = df[['Horas_estudiadas','Horas_sueno']]
y = df['Aprobo']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=LogisticRegression()
# Entrenar el modelo 
model.fit(X_train,y_train)
# Con estos se calcula la predicción
y_pred = model.predict(X_test)
```

```python
conf_matrix=confusion_matrix(y_test,y_pred) # Calcular la matriz de confusión
print(f'Matriz de confusión:\n{conf_matrix}')
```

```python
accuracy = accuracy_score(y_test,y_pred) # Calcular la precisión de la primera forma
print(f'Precisión del modelo: {accuracy*100:.2f}%')

```

```python
horas_estudiadas=6
horas_sueno=8
probabilidad_aprobar = model.predict_proba([[horas_estudiadas,horas_sueno]])[0][1]
print(f'Probabilidad de aprobar el exámen: {probabilidad_aprobar*100:.2f}%')
```
